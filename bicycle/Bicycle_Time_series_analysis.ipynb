## Getting Data From Seoul public bicycle system api

```python
import requests
from bs4 import BeautifulSoup
import json
from pprint import pprint
import numpy as np
import pandas as pd
```

```python
res = requests.get('http://openapi.seoul.go.kr:8088/5058644a58646c733132324e64486957/json/bikeList/1/1000/')
data = json.loads(res.content)
df_1 = pd.DataFrame.from_dict(data['rentBikeStatus']['row'])


res_2 = requests.get('http://openapi.seoul.go.kr:8088/5058644a58646c733132324e64486957/json/bikeList/1001/2000')
data_2 = json.loads(res_2.content)
df_2 = pd.DataFrame.from_dict(data_2['rentBikeStatus']['row'])


fin_data = pd.concat([df_1, df_2],ignore_index=True)

fin_data 
```

```python
arrr = fin_data['stationName']

arrr
```

```python
# arrr.str.split(".")

fin_data['station_No']=fin_data['stationName'].apply(lambda e: e.split('.')[0])
fin_data['station_Name']=fin_data['stationName'].apply(lambda e: e.split('.')[1])
```

```python
fin_data = fin_data.drop(columns=['stationName'])
fin_data
```

```python
colnames=['자전거번호', '대여일시', '대여대여소번호', '대여대여소명', '거치대순번', '반납일시', '반납대여소번호',
       '반납대여소명', '반납거치대순번', '이용시간', '이용거리']
```

```python
df_201901   = pd.read_csv("서울특별시 공공자전거 대여정보_201901.csv", engine='python')       # 2019_1_1 ~ 2018_1_31
df_201902   = pd.read_csv("서울특별시 공공자전거 대여정보_201902.csv", engine='python')       # 2019_2_1 ~ 2018_2_28
df_201903   = pd.read_csv("서울특별시 공공자전거 대여정보_201903.csv", engine='python')       # 2019_3_1 ~ 2018_3_31
df_201904   = pd.read_csv("서울특별시 공공자전거 대여정보_201904.csv", names=colnames, engine='python')       # 2019_4_1 ~ 2019_4_30
df_201905   = pd.read_csv("서울특별시 공공자전거 대여정보_201905.csv", names=colnames, engine='python')       # 2019_5_1 ~ 2019_5_31
```

```python
arr = fin_data['station_No']
```

```python
none_data =[]

for strarr in arr:
    if (df_201901[df_201901['대여소번호']==int(strarr)].shape[0]==0):
        none_data.append(strarr)
```

```python
fin_data = fin_data[~fin_data["station_No"].isin(none_data)]
fin_data = fin_data.reset_index(drop=True)
```

```python
d["station_id"]=arr

cols = d.columns.tolist()

cols = cols[-1:] + cols[:-1]

cols

d = d[cols]
```

```python
d
```

```python
d_in = d.copy()
d_out = d.copy()
```

```python
for arr_idx in  arr.index:
    
    sta_N = arr[arr_idx]
    
    data1_in  = df_201901[df_201901['대여소번호'] == (int(sta_N))]
    data2_in  = df_201902[df_201902['대여소번호']== (int(sta_N))]
    data3_in  = df_201903[df_201903['대여소번호']== (int(sta_N))]
    data4_in  = df_201904[df_201904['대여대여소명']== (int(sta_N))]
    data5_in  = df_201905[df_201905['대여대여소명']== (int(sta_N))]

    for i in cols[1:]:
        cnt = 0
        
        if(len(i)<7):
            str_ = "0" + i[0] + ":"
            loc_idx = i[0]
        else:
            str_ =i[:2] + ":"
            loc_idx = i[:2]
        
        data01 = (data1_in[data1_in["대여일시"].str.contains(str_)].shape[0] * 0.2)/31
        data02 = (data2_in[data2_in["대여일시"].str.contains(str_)].shape[0] * 0.2)/28
        data03 = (data3_in[data3_in["대여일시"].str.contains(str_)].shape[0] * 0.2)/31
        data04 = (data4_in[data4_in["대여일시"].str.contains(str_)].shape[0] * 0.2)/30
        data05 = (data5_in[data5_in["대여일시"].str.contains(str_)].shape[0] * 0.2)/31
        sum_ = data01 + data02 + data03 + data04 + data05
        d_in.iloc[arr_idx,(int(loc_idx)+1)] = sum_
```
