```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\19641\OSS_Project\opensourceproject\bicycle\df_bicycle.csv")
```

```python
df
```

```python
arr =  pd.read_csv(r"C:\Users\19641\OSS_Project\opensourceproject\bicycle\using_No.csv",names = ["station_No"])
```

```python
arr
```

```python
arr['station_No'] = arr['station_No'].astype(str)
```

```python
df = df[df["반납대여소번호"].isin(arr['station_No'])]
df = df.reset_index(drop=True)
df
```

```python
df = df[df["대여대여소번호"].isin(arr['station_No'])]
df = df.reset_index(drop=True)
df
```

```python
origin = df[['반납대여소번호','대여대여소번호','반납일시','대여일시']].copy()
destination = df[['반납대여소번호','대여대여소번호','반납일시','대여일시']].copy()
```

```python
origin['Path ID'] =  'No.' + origin['반납대여소번호'].map(str)+ '_' +  'No.' + origin['대여대여소번호'].map(str)
origin['Origin-Destination']='Origin'
```

```python
origin
```

```python
destination['Path ID'] = 'No.' + destination['반납대여소번호'].map(str) + '_' +  'No.' + destination['대여대여소번호'].map(str)
destination['Origin-Destination']='Destination'
```

```python
destination
```

```python
origin = origin.drop(columns=['반납대여소번호', '반납일시']).copy()
origin = origin.rename(columns={"대여대여소번호": "대여소번호", "대여일시": "일시"}).copy()
origin = origin.drop(columns=['일시']).copy()
origin
```

```python
origin['Path ID'].value_counts()
```

```python
tmp = pd.DataFrame()
tmp['Path ID'] = origin['Path ID'].value_counts().index
tmp['Count'] = origin['Path ID'].value_counts().values
tmp
```

```python
origin = origin.drop_duplicates(subset='Path ID', keep='first').reset_index(drop=True)
origin
```

```python
origin = pd.merge(origin,tmp,on='Path ID')
```

```python
destination = destination.drop(columns=['대여대여소번호', '대여일시']).copy()
destination = destination.rename(columns={"반납대여소번호": "대여소번호", "반납일시": "일시"}).copy()
destination = destination.drop(columns=['일시']).copy()
destination
```

```python
tmpp = pd.DataFrame()
tmpp['Path ID'] = destination['Path ID'].value_counts().index
tmpp['Count'] = destination['Path ID'].value_counts().values
tmpp
```

```python
destination = pd.merge(destination,tmp,on='Path ID')
```

```python
destination = destination.drop_duplicates(subset='Path ID', keep='first').reset_index(drop=True)
destination
```

```python
origin['idx'] = origin.index
destination['idx'] = destination.index
```

```python
origin
```

```python
destination
```

```python
temp = pd.concat([origin, destination], axis=0)
```

```python
temp = temp.sort_values(by=['idx'], axis=0)
```

```python
temp.reset_index(drop=True)
```

```python
temp  = temp.sort_values(['idx', 'Origin-Destination'], ascending=[True, False])
```

```python
temp = temp.reset_index(drop=True)
temp
```

```python
info = pd.read_csv(r"C:\Users\19641\OSS_Project\opensourceproject\bicycle\공공자전거 대여소 정보_201905.CSV",engine = 'python')
info = info.drop(columns=['구분','대여소명']).copy()
```

```python
info
```

```python
fin = pd.merge(temp, info, on='대여소번호')
fin = fin.sort_values(['idx', 'Origin-Destination'], ascending=[True, False])
fin = fin.drop(columns=['idx']).copy()
fin = fin.reset_index(drop=True)
fin
```

```python
arr = 'No.' + fin['대여소번호'].map(str)
fin['대여소번호'] = arr
```

```python
fin
```

```python
test = fin['Path ID'].value_counts()
tst = pd.DataFrame()
tst['Path ID'] = test.index
tst['Count'] = test.values
tst
```

```python
fin.to_csv('path.csv',encoding='utf-8-sig',index=False)
```

```python
tst.to_csv('path_count.csv',encoding='utf-8-sig',index=False)
```
